```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('car_crashes')
df.sample(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
sns.set_theme(style='whitegrid')
```

```python
fig, axes = plt.subplots(3, 2, figsize=[15, 15])
fig.suptitle("Car Crashes Dataset - Exploratory Visualizations", fontsize=16)

# histogram of total crashes
sns.histplot(df['total'], bins=15, kde=True, ax=axes[0,0])
axes[0, 0].set_title("Distribution of Total Crashes")

# box plot of insurance premium
sns.boxplot(x=df['ins_premium'], ax=axes[0, 1])
axes[0,1].set_title('Boxplot of Insurance Premium')

# Total vs Speeding
sns.scatterplot(x='speeding', y='total', data=df, ax=axes[1, 0])
axes[1, 0].set_title("Total Crashes vs Speeding")

# Bivariate analysis - Insurance Premium vs Insurance Losses
sns.scatterplot(x='ins_premium', y='ins_losses', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Insurance Premium vs Insurance Losses')


# Pairplot of select features
sns.heatmap(df[['total', 'speeding', 'alcohol', 'ins_premium', 'ins_losses']].corr(), annot=True, cmap='coolwarm', ax=axes[2,0])

# hide last subplot
axes[2, 1].axis('off')

plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()
```

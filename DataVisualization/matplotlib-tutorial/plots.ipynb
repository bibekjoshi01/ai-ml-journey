```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(50) * 100
Y_data = np.random.random(50) * 100
```

```python
plt.scatter(X_data, Y_data, c="#ff5555", marker="*", s=150)
plt.show()
```

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 86, 78, 76, 75, 74, 73, 72, 89, 90, 91]
```

```python
plt.plot(years, weights, c="r", lw=2, linestyle="--")
```

```python
x = ['C++', 'C#', 'Python', 'Java', 'Go']
y = [20, 50, 140, 1, 45]
```

```python
plt.bar(x, y, color='r', width=0.5)
```

```python
ages = np.random.normal(20, 1.5, 1000)
plt.hist(ages, bins=20, cumulative=True)
```

```python
langs = ['Python', 'C++', 'Java', 'C#', 'Go']
votes = [20, 50, 34, 12, 78]
explodes = [0, 0, 0, 0.2, 0]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", startangle=90)
```

```python
heights = np.random.normal(172, 8, 300)
plt.boxplot(heights)
```

```python
years = [2010 + i for i in range(10)]
income = [55, 67, 68, 70, 72, 75, 78, 79, 82, 85]
income_ticks = list(range(50, 80, 4))
plt.plot(years, income)
plt.title("Income of Bibek", fontname='serif')
plt.xlabel("Year")
plt.ylabel("Income in USD")
plt.yticks(income_ticks, [f"${x}" for x in income_ticks])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load iris dataaset
df = sns.load_dataset('iris')
print(df.head())
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
data = {
    'gender': ['male', 'female'] * 50,
    'education': np.random.choice(['High School', 'Bachelors', 'Masters'], size=100),
    'employment': np.random.choice(['Employed', 'Unemployed'], size=100),
    'age': np.random.randint(22, 60, size=100)
}

df = pd.DataFrame(data)
df.head()
```

```python
sns.countplot(data=df, x='gender', hue='employment')
plt.title("Gender vs Employment Status")
plt.show()
```

```python
sns.scatterplot(data=df, x='gender', y='age', hue='employment', style='employment')
plt.grid(True)
plt.show()
```

```python
x1 = np.random.normal(loc=2.0, scale=0.5, size=100)
y1 = np.random.normal(loc=2.0, scale=0.5, size=100)

x2 = np.random.normal(loc=7.0, scale=0.5, size=100)
y2 = np.random.normal(loc=3.0, scale=0.5, size=100)

x3 = np.random.normal(loc=4.0, scale=0.5, size=100)
y3 = np.random.normal(loc=7.0, scale=0.5, size=100)

df = pd.DataFrame({
    'x': np.concatenate([x1, x2, x3]),
    'y': np.concatenate([y1, y2, y3]),
    'cluster': ['A'] * 100 + ['B'] * 100 + ['C'] * 100
})
```

```python
sns.scatterplot(data=df, x='x', y='y', hue='cluster', palette='Set2')
plt.show()
```

```python
ages = np.random.normal(loc=18, scale=0.5, size=100)
marks = np.random.normal(loc=70, scale=20, size=100)

ages2 = np.random.normal(loc=22, scale=0.5, size=100)
marks3 = np.random.normal(loc=60, scale=20, size=100)

df = pd.DataFrame({
    'ages': np.concatenate([ages, ages2]),
    'marks': np.concatenate([marks, marks3]),
    'gender': ['Female'] * 42 * 2 + ['Male'] * 58 * 2,
})

sns.scatterplot(data=df, x="ages", y="marks", hue="gender")
plt.show()
```
